```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression

# Load California housing data
housing = fetch_california_housing()
X = housing.data
y = housing.target

# For visualization, select a single feature (e.g., Median Income, index 0)
X_feature = X[:, [0]]

# Fit linear regression
model = LinearRegression()
model.fit(X_feature, y)
y_pred = model.predict(X_feature)

# Plot the data and fitting line
plt.figure(figsize=(8, 6))
plt.scatter(X_feature, y, label="Actual Data", alpha=0.3)
plt.plot(X_feature, y_pred, color='red', linewidth=2, label="Fitting Line")
plt.xlabel("Median Income (10,000s USD)")
plt.ylabel("Median House Value (100,000s USD)")
plt.title("Linear Regression on California Housing (Median Income)")
plt.legend()
plt.show()
```
